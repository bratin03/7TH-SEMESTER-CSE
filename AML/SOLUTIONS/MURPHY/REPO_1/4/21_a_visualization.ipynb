# Visualization of Gaussians in exercise 21.a

```python
%matplotlib inline

from matplotlib import pyplot as plt
import numpy as np
from scipy.stats import norm

plt.rcParams['figure.figsize'] = 10, 8
```

```python
mu_1 = 0
variance_1 = 1
sigma_1 = np.sqrt(variance_1)
mu_2 = 1
variance_2 = 10 ** 6
sigma_2 = np.sqrt(variance_2)
```

# General view

```python
fig, ax = plt.subplots()
x = np.linspace(
    -4, 4, 100)
ax.plot(
    x, 
    norm.pdf(x, loc=mu_1, scale=sigma_1), 
    'r-', 
    lw=5, 
    alpha=0.6, 
    label='mu = 0, sigma = 1')
ax.plot(
    x, 
    norm.pdf(x, loc=mu_2, scale=sigma_2), 
    'g-', 
    lw=5, 
    alpha=0.6, 
    label='mu = 1, sigma = 10^3')
ax.grid()
ax.legend()
```

# Remark

The green curve is flat because the sigma_2 = 10^3 is too big for the current scale. 
In fact in the region [-4, 4], we can approximate the green curve by the tangent horizontal line in x = 0. This line is defined by y = 0.000398. Extending this line until it touches the red curve, we can make a reasanoble estimate of the points of intersection, as shown below

```python
y = norm.pdf(0, mu_2, sigma_2)
```

```python
norm.ppf(y, mu_1, sigma_1)
```

# Zoom view

Let's zoom the plot in the positive intersection

```python
fig, ax = plt.subplots()
x = np.linspace(
    3.4, 3.8, 100)
ax.plot(
    x, 
    norm.pdf(x, loc=mu_1, scale=sigma_1), 
    'r-', 
    lw=5, 
    alpha=0.6, 
    label='mu = 0, sigma = 1')
ax.plot(
    x, 
    norm.pdf(x, loc=mu_2, scale=sigma_2), 
    'g-', 
    lw=5, 
    alpha=0.6, 
    label='mu = 1, sigma = 10^3')
ax.grid()
ax.legend()
```
